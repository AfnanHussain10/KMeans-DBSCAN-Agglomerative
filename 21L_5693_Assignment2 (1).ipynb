```python
# K mean - case 1
# Code which output time, purity, distiniugable plot
inertias=[]
for i in range(1,11):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(dataset1X)
  inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
kmeans=KMeans(n_clusters=2)
start_time = time.time()
kmeans.fit(dataset1X)
end_time = time.time()
y_pred = kmeans.labels_
#finding out how accurate the model is
accuracy = accuracy_score(dataset1y, y_pred)
#TASK 2 WAS TO CALCULATE THE PURITY SO:
cluster1 = np.where(y_pred == 0)[0]
cluster2 = np.where(y_pred == 1)[0]
purity = max(np.sum(dataset1y[cluster1] == 0) + np.sum(dataset1y[cluster2] == 1), np.sum(dataset1y[cluster1] == 1) + np.sum(dataset1y[cluster2] == 0)) / len(dataset1y)
plt.scatter(dataset1X[:,0],dataset1X[:,1], c=kmeans.labels_,cmap='RdBu_r')
plt.show()
print("Time: {:.3f} seconds".format(end_time - start_time))
print("Purity: {:.3f}".format(purity))
#conclusive line
#We have good purity here with less processing time making it efficient.
```

```python
# Dbscan - case 1
# Code which output time, purity, distinguishable plot

start_time = time.time()
dbscan = DBSCAN(eps=0.2, min_samples=5)
y_pred = dbscan.fit_predict(dataset1X)
end_time = time.time()

# purity
label_true = dataset1y
label_pred = y_pred
unique_label = set(label_pred)
best_match = {}
for label in unique_label:
    best_match[label] = np.argmax(np.bincount(label_true[label_pred == label]))

y_pred_corrected = [best_match[label] for label in y_pred]
purity = accuracy_score(dataset1y, y_pred_corrected)
#results

#visualization
plt.scatter(dataset1X[:, 0], dataset1X[:, 1], c=dbscan.labels_,cmap='jet')
plt.title("DBSCAN Clustering")
plt.show()
print("Time: {:.3f} seconds".format(end_time - start_time))
print("Purity: {:.3f}".format(purity))
#conclusive line
#DBSCAN gave HIGH purity of 1 here and very less processing time of 0.2 seconds making it best for this data.
```
